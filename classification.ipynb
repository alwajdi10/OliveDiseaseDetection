```python
import os
import shutil
import glob

def clean_kaggle_working():
    """Safely deletes all files and directories in /kaggle/working except the notebook"""
    working_dir = '/kaggle/working'
    
    # Get current notebook name to preserve it
    current_notebook = glob.glob(os.path.join(working_dir, '*.ipynb'))
    
    # Delete files
    for filename in os.listdir(working_dir):
        filepath = os.path.join(working_dir, filename)
        
        # Skip current notebook files
        if filepath in current_notebook:
            continue
            
        try:
            if os.path.isfile(filepath) or os.path.islink(filepath):
                os.unlink(filepath)
            elif os.path.isdir(filepath):
                shutil.rmtree(filepath)
        except Exception as e:
            print(f"Error deleting {filepath}: {str(e)}")
    
    print("Working directory cleaned successfully!")
    print(f"Preserved notebook: {current_notebook[0] if current_notebook else 'None'}")

# Execute the cleanup
clean_kaggle_working()
```

# DATASET PREPARATION

```python
import os
import shutil
from sklearn.model_selection import train_test_split
import zipfile

# Configuration
INPUT_DIR = "/kaggle/input/olive-leaf-image-dataset/dataset"
WORKING_DIR = "/kaggle/working"
DATASET_ROOT = os.path.join(WORKING_DIR, "olive-leaf-dataset")
CLASSES = ["Healthy", "aculus_olearius", "olive_peacock_spot"]
TEST_SIZE = 0.2  # 20% for validation
SEED = 42

# Create new dataset directory in writable space
os.makedirs(DATASET_ROOT, exist_ok=True)
os.makedirs(os.path.join(DATASET_ROOT, "train"), exist_ok=True)
os.makedirs(os.path.join(DATASET_ROOT, "val"), exist_ok=True)

# Process each class
for cls in CLASSES:
    # Create class directories
    os.makedirs(os.path.join(DATASET_ROOT, "train", cls), exist_ok=True)
    os.makedirs(os.path.join(DATASET_ROOT, "val", cls), exist_ok=True)
    
    # Get all image files from input directory
    source_dir = os.path.join(INPUT_DIR, "train", cls)
    all_images = [f for f in os.listdir(source_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    
    # Split into train/val
    train_files, val_files = train_test_split(
        all_images, 
        test_size=TEST_SIZE, 
        random_state=SEED
    )
    
    # Copy train files
    for img in train_files:
        src = os.path.join(source_dir, img)
        dst = os.path.join(DATASET_ROOT, "train", cls, img)
        shutil.copy(src, dst)
    
    # Copy validation files
    for img in val_files:
        src = os.path.join(source_dir, img)
        dst = os.path.join(DATASET_ROOT, "val", cls, img)
        shutil.copy(src, dst)

print("Dataset preparation complete!")
print(f"Dataset copied to: {DATASET_ROOT}")
print(f"Train samples: {len(train_files)} per class")
print(f"Validation samples: {len(val_files)} per class")
```

```python
from ultralytics import YOLO
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import os

# Configuration
TEST_SET_PATH = "/kaggle/input/olive-leaf-image-dataset/dataset/test"
CLASSES = ["Healthy", "aculus_olearius", "olive_peacock_spot"]

# Load both models for comparison
original_model = YOLO("/kaggle/input/classifer/pytorch/default/1/best.pt")
improved_model = YOLO("/kaggle/working/healthy-refinement/final_training/weights/best.pt")

def generate_confusion_matrix(model, model_name):
    """Generate and plot confusion matrix for a model"""
    true_labels = []
    pred_labels = []
    
    for class_name in CLASSES:
        class_dir = os.path.join(TEST_SET_PATH, class_name)
        for img_file in os.listdir(class_dir):
            img_path = os.path.join(class_dir, img_file)
            results = model(img_path, verbose=False)
            true_labels.append(class_name)
            pred_labels.append(CLASSES[results[0].probs.top1])
    
    # Generate confusion matrix
    cm = confusion_matrix(true_labels, pred_labels, labels=CLASSES)
    
    # Plot confusion matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=CLASSES, 
                yticklabels=CLASSES)
    plt.title(f'Test Set Confusion Matrix ({model_name})', fontsize=16)
    plt.xlabel('Predicted Label', fontsize=14)
    plt.ylabel('True Label', fontsize=14)
    plt.tight_layout()
    plt.savefig(f'{model_name}_confusion_matrix.png')
    plt.show()
    
    # Calculate class accuracies
    class_acc = {}
    for i, cls in enumerate(CLASSES):
        class_acc[cls] = cm[i,i] / cm[i].sum()
    
    # Calculate overall accuracy
    overall_acc = np.trace(cm) / np.sum(cm)
    
    return cm, class_acc, overall_acc

# Generate confusion matrices for both models
print("Evaluating original model...")
orig_cm, orig_class_acc, orig_overall_acc = generate_confusion_matrix(original_model, "Original")

print("\nEvaluating improved model...")
imp_cm, imp_class_acc, imp_overall_acc = generate_confusion_matrix(improved_model, "Improved")

# Print comparison
print("\nModel Comparison:")
print(f"{'Metric':<25} | {'Original':<10} | {'Improved':<10} | {'Change':<10}")
print("-" * 65)
print(f"{'Overall Accuracy':<25} | {orig_overall_acc:.4f}    | {imp_overall_acc:.4f}    | {imp_overall_acc - orig_overall_acc:+.4f}")

for cls in CLASSES:
    print(f"{cls + ' Accuracy':<25} | {orig_class_acc[cls]:.4f}    | {imp_class_acc[cls]:.4f}    | {imp_class_acc[cls] - orig_class_acc[cls]:+.4f}")

# Visual improvement on problematic images
problematic_images = [
    "/kaggle/input/olive-leaf-image-dataset/dataset/test/Healthy/IMG_20190806_164445.jpg",
    "/kaggle/input/olive-leaf-image-dataset/dataset/test/Healthy/IMG_20190729_170325.jpg"
]

plt.figure(figsize=(15, 10))
for i, img_path in enumerate(problematic_images):
    img = plt.imread(img_path)
    
    # Original model prediction
    orig_results = original_model(img_path, verbose=False)
    orig_pred = orig_results[0].names[orig_results[0].probs.top1]
    orig_conf = orig_results[0].probs.top1conf
    
    # Improved model prediction
    imp_results = improved_model(img_path, verbose=False)
    imp_pred = imp_results[0].names[imp_results[0].probs.top1]
    imp_conf = imp_results[0].probs.top1conf
    
    # Display side-by-side comparison
    plt.subplot(2, 2, i*2+1)
    plt.imshow(img)
    plt.title(f"Original: {orig_pred} ({orig_conf:.2f})", 
             color='red', fontsize=12)
    plt.axis('off')
    
    plt.subplot(2, 2, i*2+2)
    plt.imshow(img)
    plt.title(f"Improved: {imp_pred} ({imp_conf:.2f})", 
             color='green' if imp_pred == "Healthy" else 'red', 
             fontsize=12)
    plt.axis('off')

plt.tight_layout()
plt.savefig('problematic_images_comparison.png')
plt.show()
```

```python
from ultralytics import YOLO
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import torch

# Paths
model_path = '/kaggle/input/classify1/pytorch/default/1/classify1.pt'
test_dir   = '/kaggle/input/olive-leaf-image-dataset/dataset/test/'

# Data transform — no Normalize, let Ultralytics handle it
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

# Data loader
dataset     = datasets.ImageFolder(test_dir, transform=transform)
loader      = DataLoader(dataset, batch_size=32, shuffle=False)
class_names = dataset.classes

# Load model
model = YOLO(model_path)

# Inference (no verbose prints)
all_preds  = []
all_labels = []

for images, labels in loader:
    # pass verbose=False to suppress per-image logs
    results = model(images, verbose=False)  
    preds   = torch.tensor([r.probs.top1 for r in results])
    all_preds.extend(preds.cpu().numpy())
    all_labels.extend(labels.numpy())

# Build and plot confusion matrix
cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d',
            xticklabels=class_names,
            yticklabels=class_names,
            cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
